## Task 1 Workflow & Methodology

The foundation of this project relies on a structured analytical workflow designed to handle the high volatility and non-linear nature of oil prices.

### Workflow Steps:
1. **Data Ingestion**: Loading raw daily Brent prices and handling inconsistent date formats (e.g., Apr 22, 2020 vs ISO formats).
2. **Stationarity Analysis**: Using the Augmented Dickey-Fuller (ADF) test to identify if the series can be modeled directly or needs transformation.

4. **Volatility Assessment**: Using rolling metrics to identify periods of localized instability.

### Key Assumptions:
- **Efficiency**: We assume that major global events are reflected in price changes shortly after occurrence (Market Efficiency Hypothesis).
- **Data Reliability**: The World Bank/Energy Information Administration data is assumed to be the gold standard for historical daily settlement prices.
- **Logarithmic Returns**: We assume that price changes are multiplicative, making log-returns the most appropriate stationary transformation for modeling.

## 3. Stationarity Testing
We use the Augmented Dickey-Fuller (ADF) test to check for stationarity.
Null Hypothesis (H0): The time series has a unit root (is non-stationary).
Alternate Hypothesis (H1): The time series is stationary.


```python
print("ADF Test for Raw Prices:")
check_stationarity(df['Price'])
```

If the raw prices are non-stationary (p-value > 0.05), we calculate log returns to achieve stationarity.
$$ r_t = \ln(P_t) - \ln(P_{t-1}) $$


```python
# Calculate Log Returns
df = calculate_log_returns(df)

print("ADF Test for Log Returns:")
check_stationarity(df['Log_Returns'])

plot_time_series(df, 'Log_Returns', 'Brent Oil Price Log Returns', 'Log Return', color='green')
```
